```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# loading csv's as dataframes

# bank df
bank = pd.read_csv('data/bank.csv')[['Rohan', 'Ethan', 'Yarden', 'Quota', 'Bit Avg. Sum']]
bank = bank[bank['Rohan'].notna()]  # drop rows that have yet to be filled in

# schedule df
sched = pd.read_csv('data/schedule.csv')

# values df
vals = pd.read_csv('data/values.csv')[['Chore', 'Frequency', 'Nuggies']].dropna()
```

```python
# graphing the bank
rohan = bank['Rohan']
ethan = bank['Ethan']
yarden = bank['Yarden']
quota = bank['Quota']
bits = bank['Bit Avg. Sum']

# only bank accounts
plt.plot(rohan, label='rohan')
plt.plot(ethan, label='ethan')
plt.plot(yarden, label='yarden')
plt.legend()
plt.title('bank accounts')
plt.xlabel('week')
plt.ylabel('nuggies')
plt.show()

# bank accounts with quota and SAB (Sum of Average Bits)
figure(figsize=(8, 6), dpi=80)
plt.plot(rohan, label='rohan')
plt.plot(ethan, label='ethan')
plt.plot(yarden, label='yarden')
plt.plot(quota, label='quota')
plt.plot(bits, label='SAB')
plt.legend()
plt.title('bank accounts w/ quota + SAB')
plt.xlabel('week')
plt.ylabel('nuggies')
plt.show()
```
